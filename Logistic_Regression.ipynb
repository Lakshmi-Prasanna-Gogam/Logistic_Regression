**Regression:**

Regression is a supervised machine learning technique where the model predicts continuous values based on the labeled data

**Logistic Regression:**

A type of regression which is used for binary classification.

**Objective**: Build a binary classifier using logistic regression.

**Tools**: Scikit-learn, Pandas, Matplotlib.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Data Collection and Preprocessing**

```python
dataset = pd.read_csv('/content/data.csv')
dataset
```

**Handling missing values using mean/median/dropna()/fillna()**

```python
dataset
```

```python
dataset.dropna()
dataset.fillna(0)
```

**Converting categorical data into numerical data (Feature scaling)**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
catg_cols = dataset.select_dtypes(include = ['object','category']).columns
catg_cols
```

```python
dataset['diagnosis'] = le.fit_transform(dataset['diagnosis'])
```

**Generating summary statistics**

```python
x = dataset.drop('diagnosis',axis=1)
```

```python
y = dataset[['diagnosis']]
```

```python
dataset.describe()
```

**Split data into train-test sets**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

**Training dataset**

```python
x_train
```

```python
y_train
```

**Testing dataset**

```python
x_test
```

```python
y_test
```

**Fit a Logistic Regression model using sklearn.linear_model.**

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

**Predicted values on testing dataset**

```python
model.predict(x_test)
```

**Evaluate with confusion matrix, precision, recall, ROC-AUC.**

```python
from sklearn.metrics import precision_score,recall_score,confusion_matrix,roc_auc_score
```

**Precision score on testing data**

```python
precision = precision_score(y_test,model.predict(x_test))
precision
```

**Recall score on testing data**

```python
recall = recall_score(y_test,model.predict(x_test))
recall
```

**Confusion matric on testing data**

```python
cm = confusion_matrix(y_test,model.predict(x_test))
cm
```

**ROC-AUC score on testing data**

```python
rocauc = roc_auc_score(y_test,model.predict(x_test))
rocauc
```

**Tune threshold and explain sigmoid function.**

```python
model.coef_  #Coefficients
```

```python
model.intercept_ #intercept
```

```python
from sklearn.metrics import classification_report
```

```python
threshold = 0.3
y_pred_custom = (model.predict(x_test) >= threshold).astype(int)
```

```python
class_report = classification_report(y_test,y_pred_custom)
print(class_report)
```
